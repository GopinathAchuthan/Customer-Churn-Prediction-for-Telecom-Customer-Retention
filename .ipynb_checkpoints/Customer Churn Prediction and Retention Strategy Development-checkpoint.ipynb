# **Customer Churn Prediction for Telecom Customer Retention**

# Problem Statement
---

**Problem Definition:**
Customer churn is a critical issue for telecom companies, as retaining existing customers is often more cost-effective than acquiring new ones. In this project, we aim to develop a machine learning model capable of predicting whether a customer will churn or remain with the telecom service based on various customer attributes. By accurately predicting churn, the company can focus on targeted retention strategies to minimize customer loss.

**Objective:**
The goal is to identify customers who are at risk of churning, based on a variety of customer-related data. With this predictive model, the company can proactively engage with customers who are likely to leave and implement retention programs to reduce churn.

**Dataset:** Telco-Customer-Churn.csv

**Classification Problem:**

This is a **binary classification problem** where we are predicting whether a customer has churned (left) or not within the last month.

- **Target Variable (Dependent Variable):** 
  - **Churn** (1: Customer churned, 0: Customer did not churn)
  
- **Features (Independent Variables):**
  - **Services Used:** phone, multiple lines, internet, online security, online backup, device protection, tech support, streaming TV and movies.
  - **Customer Account Information:** tenure (length of time as a customer), contract type, payment method, paperless billing, monthly charges, total charges.
  - **Demographic Information:** gender, age range, presence of partners, and dependents.

**Classification Type:**
- **Binary Classification:** We are predicting whether the customer will churn (1) or not (0).

**Success Criteria:**

- **Recall (Sensitivity):** Proportion of actual churned customers correctly identified. High recall reduces false negatives, ensuring churned customers are captured.
- **F1-Score:** Balances precision and recall, essential for imbalanced datasets. It ensures a good trade-off between identifying churn and minimizing false positives.
- **AUC-ROC:** Measures the model's ability to distinguish between churned and non-churned customers across thresholds. A higher AUC reflects better model performance.
  
We aim to build a model that will help the company identify at-risk customers and implement more targeted retention strategies.

# Model Selection
---

Given the characteristics of the datasetâ€”7032 rows, 3 numeric columns, 10 categorical columns, and an imbalanced target variable (Churn), selecting appropriate models is key to handling class imbalance, the mix of feature types, and achieving high performance.

## Possible Models

### **1. Logistic Regression with Class Weights**
- **Why:** A simple, interpretable linear model, suitable for binary classification. Adjusting `class_weight='balanced'` helps address class imbalance.
- **Advantages:** 
  - Fast to implement.
  - Provides probabilities for threshold tuning.
  - Interpretable and easy to understand.
- **Considerations:** Assumes linear relationships, limiting its ability to capture complex patterns.

### **2. Decision Trees**
- **Why:** Simple model that handles both numerical and categorical data. Can address class imbalance with `class_weight='balanced'`.
- **Advantages:**
    - Easy to interpret and visualize.
    - Can handle both numerical and categorical data.
    - Non-linear relationships between features can be captured.
    - No need for feature scaling or normalization.
- **Considerations:**
    - Prone to overfitting, especially with small or noisy datasets.
    - Less robust to outliers.
    - Requires pruning or tuning to avoid overfitting.
### **3. Random Forest Classifier**
- **Why:** An ensemble model that combines decision trees, ideal for handling both numeric and categorical features (after encoding). It can model complex data relationships and address class imbalance with `class_weight='balanced'`.
- **Advantages:** 
  - Captures non-linear patterns.
  - Handles both feature types efficiently.
  - Robust to overfitting.
- **Considerations:** Computationally expensive and harder to interpret than simpler models.

### **4. Gradient Boosting (XGBoost, LightGBM, or CatBoost)**
- **Why:** Powerful ensemble methods that perform well with imbalanced data and complex relationships. LightGBM and CatBoost handle categorical features efficiently.
- **Advantages:**
  - Excellent for imbalanced datasets.
  - Robust to overfitting.
  - CatBoost handles categorical features natively, reducing the need for encoding.
- **Considerations:** Requires careful hyperparameter tuning and is more computationally intensive.

### **5. Neural Networks (Multi-layer Perceptron)**
- **Why:** Capable of modeling complex, non-linear relationships between features. Useful for intricate patterns in data.
- **Advantages:** 
  - Models non-linear relationships well.
  - Powerful for learning complex patterns.
- **Considerations:** Requires extensive tuning and a sufficiently large dataset. Computationally expensive and prone to overfitting without regularization.

### Note:
- **No Encoding Needed:** CatBoost (handles categorical features natively).
- **Needs Encoding:** XGBoost, LightGBM, Logistic Regression, Random Forest, Neural Networks (require categorical variables to be encoded).
- **No Scaling Needed:** XGBoost, CatBoost, LightGBM, Random Forest (tree-based models don't require scaling).
- **Scaling Needed:** Logistic Regression, Neural Networks (these algorithms are sensitive to the scale of the features).

## Model Evaluation and Performance Visualization

```python
def evaluation_metrics(y_test, y_pred, y_prob):
    # Compute evaluation metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob)
    
    # Compute the confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    cm_df = pd.DataFrame(cm, columns=['Predicted No', 'Predicted Yes'], index=['Actual No', 'Actual Yes'])
    
    # Print the results with 4 decimal places
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-score: {f1:.4f}")
    print(f"AUC: {auc:.4f}")
    print("\nConfusion Matrix:")
    print(cm_df)
```

```python
def display_ROC_curve(y_test, y_prob):
    # ROC Curve (Visual representation)
    fpr, tpr, thresholds = roc_curve(y_test, y_prob)

    # Calculate ROC AUC
    roc_auc = roc_auc_score(y_test, y_prob)
    
    plt.figure(figsize=(5, 4))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.4f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc="lower right")
    plt.show()
```

```python
def display_Precision_Recall_curve(y_test, y_prob):
    # Calculate precision and recall for different thresholds
    precision, recall, thresholds = precision_recall_curve(y_test, y_prob)
    
    # Plot the Precision-Recall curve
    plt.figure(figsize=(5, 4))
    plt.plot(recall, precision, marker='.')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision-Recall Curve')
    plt.show()
```

## Logistic Regression: HyperParameter Tuning

In this section, we will tune the hyperparameters of the Logistic Regression using GridSearchCV to find the optimal model

```python
param_grid = [
    {'penalty': [None],
     'solver': ['lbfgs', 'newton-cg', 'sag', 'saga'],
     'max_iter': [1000, 5000],
     'class_weight': [None, 'balanced']
    },
    {'penalty': ['l1'],
     'solver': ['liblinear', 'saga'],
     'C': [0.01, 0.1, 1, 10],
     'max_iter': [1000, 5000],
     'class_weight': [None, 'balanced']
    },
    {'penalty': ['l2'],
     'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
     'C': [0.01, 0.1, 1, 10],
     'max_iter': [1000, 5000],
     'class_weight': [None, 'balanced']
    },
    {'penalty': ['elasticnet'],
     'solver': ['saga'],
     'C': [0.01, 0.1, 1, 10],
     'max_iter': [1000, 5000],
     'class_weight': [None, 'balanced'],
     'l1_ratio': [0, 0.5, 1],
    }
]
```

```python
# Initialize the Logistic Regression model
model = LogisticRegression()

grid_search = GridSearchCV(estimator=model, 
                           param_grid=param_grid, 
                           cv=5, 
                           scoring='recall',  # maximize recall for churn
                           verbose=1, 
                           n_jobs=-1
                          )

grid_search.fit(X_train, y_train)
```

```python
# Print the best parameters and the corresponding accuracy
print("Best Parameters: ", grid_search.best_params_)
print("Best Recall: ", grid_search.best_score_)
```

```python
# Evaluate the best model on the test set
best_model = grid_search.best_estimator_

# check the coefficients and intercept of the best logistic regression model
print("Model Coefficients: ", best_model.coef_)
print("Model Intercept: ", best_model.intercept_)
```

```python
# Make predictions on the test data
y_pred_best = best_model.predict(X_test)
y_prob_best = best_model.predict_proba(X_test)[:, 1]  # Probability predictions of the positive class
```

```python
# Call the evaluation function
print('Logistic Regression Evaluation Metrics:') 
evaluation_metrics(y_test, y_pred_best, y_prob_best)
```

# Conclusion
---

- **Best for minimizing false negatives (high recall)**: **XGBoost**
- **Best overall balanced model**: **LightGBM**
- **Good general-purpose model**: **Random Forest**

In this project, the goal is to predict customer churn and proactively target retention efforts to minimize churn for a telecom company. Based on the evaluation of different machine learning models, the following conclusions can be drawn:

- **XGBoost** is the best model for minimizing false negatives (high recall). Its ability to correctly identify churned customers is the highest among all models, making it an ideal choice if the primary objective is to minimize missed churn predictions. However, it has a lower precision, so further tuning or threshold adjustments may be necessary to improve overall performance.

- **LightGBM** is the best overall balanced model. It provides a strong recall (good at identifying churned customers), a competitive F1-score, and high AUC-ROC, indicating a well-rounded performance. Given its balance between recall and precision, along with its robust AUC-ROC score, LightGBM offers a good trade-off for identifying at-risk customers while avoiding an overwhelming number of false positives.

- **Random Forest** is a good general-purpose model, delivering a reasonable balance between recall, precision, and F1-score. While it may not excel in terms of recall like XGBoost or LightGBM, it provides stable performance across various metrics and could serve as a reliable baseline.

Given the project's objective of proactive retention, **LightGBM** is recommended as the best starting point. It strikes a good balance between recall and precision, which is crucial for identifying customers at risk of churning while minimizing false positives. However, if the goal is to capture as many churned customers as possible, **XGBoost** should also be explored further, especially with threshold tuning to improve its precision.

In summary, **LightGBM** offers a solid foundation for building an effective churn prediction model, but **XGBoost** can be considered if the focus shifts towards identifying every churned customer, even at the cost of precision.
